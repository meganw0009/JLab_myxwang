```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize, least_squares, curve_fit
from scipy.stats import poisson, chi2, f, mode
```

```python
# Summing the cobalt spectra
summed_data = np.zeros(2048)

for i in range(271):
    co_spectra = f"cobalt_long_integ\\cobalt_snapshot_{i}_spectra.Spe"
    data = np.genfromtxt(co_spectra, delimiter='\n', skip_header=12)
    for elt in range(2048):
        summed_data[elt] += data[elt]

# data = np.genfromtxt(co_spectra, delimiter='\n', skip_header=12)

plt.bar(range(650,750),summed_data[650:750], color = '#1E88E5')

# savename = "cobalt_long_integ\\cobalt_summed_spectra.Spe"
# np.savetxt(savename, summed_data, delimiter='\n')
```
